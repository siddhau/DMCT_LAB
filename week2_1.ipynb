```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
BikeShare = pd.read_csv('bikeSharingHourly.csv',index_col='instant',parse_dates=[2])
BikeShare
```

```python
BikeShare.boxplot(column=['registered','casual'],by='weekday',figsize=(18,8))
```

```python
df5 = BikeShare.groupby('weekday')
```

```python
df5['registered'].mean()
df6 = pd.concat([df5['registered'].mean(),df5['casual'].mean()],axis=1)
```

```python
df6.head()
```

```python
df6.plot.bar(figsize=(8,8))
```
