```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.genfromtxt('iris_numeric.txt',delimiter=',');
data
```

```python
mn=np.mean(data,axis=0)
mn[4]=0
data=data-mn
data
```

```python
N=len(data)
temp=np.random.permutation(range(N))
test=[]
train=[]
for i in range(N):
    if i<(0.75*N):
        train.append(data[temp[i]])
    else:
        test.append(data[temp[i]])
train=np.array(train)
test=np.array(test)
print(len(train),len(test))
```

```python
ytrain=train[:,4]
ytest=test[:,4]
train=train[:,:4]
test=test[:,:4]
ypred=[-1 for x in range(len(test))]
K=1
buff=[]
while K<50: #odd
    for i in range(len(test)):
        temp=train-test[i]
        temp=temp**2
        dis=np.sum(temp,axis=1)
        dis=np.argsort(dis)
        dis=dis[:K]
        freq=[0 for x in range(3)]
        mx=0
        id1=-1
        for j in range(K):
            yo=int(ytrain[dis[j]]-1)
            freq[yo]=freq[yo]+1
            if freq[yo]>mx:
                mx=freq[yo]
                id1=yo
        ypred[i]=id1+1
    result=ypred==ytest
    #print(result)
    acc=(np.count_nonzero(result)/len(result))
    acc*=100
    buff.append(acc)
    K+=1
plt.plot(range(len(buff)),buff)
plt.show()
    
```
