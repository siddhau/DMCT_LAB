```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.genfromtxt('regression_data.csv',delimiter=',')
```

```python
mn=np.mean(data,axis=0)
sd=np.std(data,axis=0)
data=(data-mn)/sd
a=data[:,0]
b=data[:,1]
plt.scatter(a,b)
plt.show()
cov=(sum(a*b))/(len(a))
pearl=cov/(sd[0]*sd[1])
print(pearl,cov)
```

```python
N=len(a)
temp=np.random.permutation(range(N))
train=[]
test=[]
for i in range(N):
    if i<(75*N/100):
        train.append(data[temp[i]])
    else :
        test.append(data[temp[i]])
train=np.array(train)
test=np.array(test)
print(len(train),len(test))
```

```python
N=len(train)
a=train[:,0]
b=train[:,1]
mn=np.mean(train,axis=0)
std=np.std(train,axis=0)
w1=(np.mean(a*b)-mn[0]*mn[1])/((np.mean(a*a)-mn[0]*mn[0]))
w0=mn[1]-w1*mn[0]
b1=[]
for i in range(N):
    b1.append(w0+w1*a[i])
plt.scatter(a,b1)
plt.scatter(a,b)
plt.show()
```

```python
y=train[:,1:2]
train=train[:,0:1]
train=np.hstack((np.ones((N,1)),train))
temp1=np.zeros((N,1))
temp2=np.zeros((N,1))
for i in range(N):
    temp1[i][0]=train[i][1]**2
    temp2[i][0]=train[i][1]**3
train=np.hstack((train,temp1,temp2))
print (train)
```

```python
alpha=0.1
gen=100
N,M=train.shape
theta=np.zeros((1,M))
print(N,M)
for i in range(gen):
    h=np.dot(train,np.transpose(theta)) #N*1
    temp=np.dot(np.transpose(h-y),train)
    theta=theta-(alpha/N)*temp
h=np.dot(train,np.transpose(theta))
x=train[:,1:2]
plt.scatter(x,h)
plt.scatter(x,y)
plt.show()
```
