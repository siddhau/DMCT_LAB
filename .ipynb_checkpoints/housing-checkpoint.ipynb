```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
inp=pd.read_csv('housing.csv')
inp.info() 
inp.shape   # give dimensions
inp.describe()  # give statistical description
```

```python
x=inp.head(5) # returns data frame of first 5 rows
inp['ocean_proximity'].unique()  #give unique value
inp['ocean_proximity'].nunique()  #give no of unique value
inp.hist(column='median_income',bins=10)
```

```python
inp.plot.scatter(x='longitude',y='latitude')
```

```python
cool=inp.groupby('ocean_proximity')
cool.groups
cool1=cool['population'].mean()
cool1.plot.bar()
```

```python
#inp=inp.iloc[0:50]
cool=inp.groupby('median_house_value');
x=cool['total_rooms'].mean()
x=pd.DataFrame(x)
x.reset_index(inplace=True)
x.columns=['price','avg rooms']
x.plot.scatter(x='price',y='avg rooms')
```

```python
size=inp['population']/25
inp.plot.scatter(x='longitude',y='latitude',c='median_house_value',figsize=(10,11),s=size,alpha=0.1,cmap=plt.get_cmap('jet'))
```
